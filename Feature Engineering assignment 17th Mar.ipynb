```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
#M1 Dropping the rows
df_Cleaned = df.dropna()
df_Cleaned.isnull().sum()
```

```python
df.shape,df_Cleaned.shape
```

```python
#M2 Imputation with mean
df[df.isnull().any( axis = 1)]
```

```python
df.isnull().any(axis = 0)
```

```python
df.isnull().sum()
```

```python
df['mean_Age'] = df['age'].fillna(df['age'].mean())
df['median_Age'] = df['age'].fillna(df.age.median())
df['mode_age']= df['age'].fillna(df['age'].mode().iloc[0]) 
df['forward_fill'] = df['age'].fillna(method = 'ffill')
df['backward_fill'] = df['age'].fillna(df['age'].bfill())
```

```python
df
```

```python
df['linear_interpolation'] = df['age'].interpolate(method='linear') 
df['quadratic_interpolation'] = df['age'].interpolate(method='quadratic') 
```

```python
Imbalnced data is where classes have different proportions.

A classification data set with skewed class proportions is called imbalanced. 
Classes that make up a large proportion of the data set are called majority classes. Those that make up a smaller proportion are minority classes.

Consider the following example of a model that detects fraud. Instances of fraud happen once per 200 transactions in this data set, 
so in the true distribution, about 0.5% of the data is positive.

Why would this be problematic? With so few positives relative to negatives, the training model will spend most of its time on negative examples and not learn enough from positive ones. 
For example, if your batch size is 128, many batches will have no positive examples, so the gradients will be less informative.
```

```python
Upsampling is the process of randomly duplicating observations from the
minority class to reinforce its signal.

Downsampling involves randomly removing observations from the majority
class to prevent its signal from dominating the learning algorithm.
```

```python
Data Augmentation is artifically increasing the training data by creating modified copies of a dataset using existing data.
 It includes making minor changes to the dataset or using deep learning to generate new data points.
```

```python
Synthetic data is generated artificially without using the original dataset. It often uses DNNs (Deep Neural Networks) and GANs (Generative Adversarial Networks) to generate synthetic data.

Note: the augmentation techniques are not limited to images. You can augment audio, video, text, and other types of data too.

When Should You Use Data Augmentation?

To prevent models from overfitting.
The initial training set is too small.
To improve the model accuracy.
To Reduce the operational cost of labeling and cleaning the raw dataset.
```

```python
Limitations of Data Augmentation
The biases in the original dataset persist in the augmented data.
Quality assurance for data augmentation is expensive.
Research and development are required to build a system with advanced applications. For example, generating high-resolution images using GANs can be challenging.
Finding an effective data augmentation approach can be challenging.
```

```python
SMOTE : synthetic minority oversampling techniques.

SMOTE is specifically designed to tackle imbalanced datasets by generating synthetic samples for the minority class.
```
